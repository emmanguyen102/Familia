# 2. Opiskelijoiden analyysi kotimaidensa perusteella

```python
data["Kotimaa"] = data.Kotimaa.str.strip()

for value in data["Kotimaa"].values:
    if value in {"Russia","Finland","Helsinki"}:
        data.Kotimaa.replace({"Russia":"Venäjä", "Finland":"Suomi", "Helsinki": "Suomi"}, inplace=True)

kotimaa = data.groupby("Kotimaa")["ID"].count().sort_values(ascending=False)
kotimaa.head(25)
```
